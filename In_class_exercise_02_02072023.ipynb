Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):


  
import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd

pages = int(input("Enter num of pages you want to process: "))

amazon_reviews = []
for i in range(pages):
  url = f"https://www.amazon.com/s?k=watches+for+women&page={i}"
  headers = {
      'authority': 'www.amazon.com',
      'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
      }
  amazon_data = requests.get(url, headers=headers)
  soup = BeautifulSoup(amazon_data.content, "html.parser")
  products_data = soup.findAll("span", attrs={"class": "a-size-base-plus a-color-base a-text-normal"})
  j = 0
  for data in products_data:
    title = data.text
    url = "https://www.amazon.com"+str(data.parent.get("href"))
    # print(url)
    # print(str(data.parent.parent.parent.find_next_sibling("div").div.span.text).split())
    div_ele = data.parent.parent.parent.find_next_sibling("div")
    if not div_ele:
      continue
    div_ele = div_ele.find("span", attrs={"class": "a-icon-alt"})
    if not div_ele:
      continue
    rating = str(div_ele.text)
    # print(rating)
    rating = float(rating.split()[0])
    amazon_reviews.append([title, url, rating])
  time.sleep(5)
  print(f"Processed page: {(i+1)}")

amazon_Df = pd.DataFrame(amazon_reviews, columns=["Title", "UrL", "Ratings"])
print(len(amazon_Df))
amazon_Df.head()
```
