(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here
import random
from statistics import mean

random_list=[]
for i in range(12):
  random_list.append(random.randint(1,100))
print(random_list)
random_list.sort()
print("the ascending order of the list :",random_list)
random_list.sort(reverse=True)
print("the decesnding order of the list :",random_list)
average= mean(random_list)
print("the average of the list:", round(average,2))



```

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```python
# Your code here
import math
def vol():
  for r in range(1,11):
    yield f"{r} \t {(4/3)*(math.pi)*(r**3)}"
print("r \t V")
for i in vol():
  print(i)
```

Write a program to implement the following questions.

```python
# Your code here

import pandas as pd
df = pd.read_csv(r"Assignment1_denton_housing.csv")
df.head()




```

```python
df
```

```python
ohu_df  = df[df["title_field"] == "Occupied Housing Units"]
ohu_df = ohu_df.sort_values("year").reset_index(drop=True)
ohu_df
```

```python
# Calculating the difference in the number of Occupied Housing Units from year to year 

print("Year period \t housing unit diff")
udiff_map = {}
for row in range(1, len(ohu_df)):
  value_chng = ohu_df.loc[row, 'value'] - ohu_df.loc[row - 1, 'value']
  udiff_map[f"{ohu_df.loc[row - 1, 'year']} -> {ohu_df.loc[row, 'year']}"] = value_chng

udiff_map = dict(sorted(udiff_map.items(), key = lambda x: x[1]))
for key, value in udiff_map.items():
  print(f"{key} \t {value}")
```

```python
# For all the years, calculate the percentage of housing units which are vacant an occupied.

vac_df, occ_df = df[(df["title_field"] == "Vacant Housing Units")].reset_index(drop=True), df[(df["title_field"] == "Occupied Housing Units")].reset_index(drop=True)
vac_df = vac_df.sort_values("year").reset_index(drop=True)
vac_df
```

```python
occ_df = occ_df.sort_values("year").reset_index(drop=True)
occ_df
```

```python
print("Year \t Vacant Housing Units \t Occupied Housing Units") 
for i in range(len(vac_df)):
  print(f"{vac_df.loc[i, 'year']} \t {round((vac_df.loc[i, 'value']/(vac_df.loc[i, 'value']+ occ_df.loc[i, 'value']))*100, 2)}%\t \t \t {round((occ_df.loc[i, 'value']/(vac_df.loc[i, 'value']+ occ_df.loc[i, 'value']))*100,2)}%")
```

```python
# Calculate and print the valued and years in which the highest number of housing units were vacant and occupied

highest_vac_id, highest_occ_idx = vac_df['value'].argmax(), occ_df['value'].argmax()
print("\t \t \t Year \t Value")
print(f"Vacant Housing Units \t {vac_df.loc[highest_vac_id, 'year']} \t {vac_df.loc[highest_vac_id, 'value']}")
print(f"Occupied Housing Units \t {occ_df.loc[highest_occ_idx, 'year']} \t {occ_df.loc[highest_occ_idx, 'value']}")
```

```python
# Calculate the harmonic mean of the total housing units
def harmonic_mean(df):
  toth_units = df[df["title_field"] == "Total Housing Units"]
  toth_units_values = list(toth_units.value)
  sum_toth_unit = 0
  for x in toth_units_values:
    sum_toth_unit += (1/x)
  return len(toth_units_values)/sum_toth_unit
print(f"Harmonic Mean of the total housing units: {harmonic_mean(df)}")
```
